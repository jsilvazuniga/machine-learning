```python
# Modeling a Bank account using OOP
class BankAccount:
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance
        
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. New balance: ${self.balance}")
        else:
            print("Deposit amount must be positive.")
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds.")
        elif amount <= 0:
            print("Withdrawal amount must be positive.")
        else:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")
    
    def get_balance(self):
        return f"Current balance: ${self.balance}"
```

```python
# create an account
account = BankAccount("John Doe", 1000) 
```

```python
account.get_balance()
```

```python
account.deposit(500)
```

```python
account.withdraw(150)
```

# Inheritance

```python
# Parent and child classes
class Car:
    def __init__(self, windows, doors, engineType):
        self.windows = windows
        self.doors = doors
        self.engineType = engineType

    def drive(self):
        print(f"The person is driving the {self.engineType} car.")

        
    
```

```python
car1 = Car(4, 4, "petrol")
car1.drive()
```

```python
class Tesla(Car):
    def __init__(self, windows, doors, engineType, is_Self_driving):
        super().__init__(windows, doors, engineType)
        self.is_Self_driving = is_Self_driving

    def drive(self):
        if self.is_Self_driving:
            print(f"The person is driving the {self.engineType} Tesla autonomously.")
        else:
            print(f"The person is driving the {self.engineType} Tesla manually .")
```

```python
car2 = Tesla(4, 4, "electric", True)
car2.drive()
```

# Multiple Inheritance

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("Subclass must implement this method.")

class Pet:
    def __init__(self, owner):
        self.owner = owner

    def speak(self):
        self.owner = owner

class Dog(Animal, Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)

    def speak(self):
        print(f"{self.owner} is owned to {self.name}.") 
        print(f"{self.name} says Woof!")
```

```python
dog1 = Dog("Buddy", "Alice")
dog1.speak()
```

# Polymorphism

Method overriding allows a child class to provide a specific implementation of the mothod that is already defined in its parent class

```python
# base class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("Sound of the animal.")

# derived class
class Dog(Animal):
    def speak(self):
        print(f"{self.name} says Woof!")

# derived class 2
class Cat(Animal):
    def speak(self):
        print(f"{self.name} says Meow!")
```

```python
dog = Dog("Rabolio")
cat = Cat("Nuna")

dog.speak()
cat.speak()
```

```python
def animal_speak(animal):
    animal.speak()
    
animal_speak(dog)
```

```python
# Polimorphism with functions and methods
class Shape:
    def area(self):
        print("Calculating area of the shape.")
        
# Derived class 1
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
        
    def area(self):
        return self.width * self.height
    
# Derived class 2
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
        
    def area(self):
        import math
        return math.pi * (self.radius ** 2) 
```

```python
def print_area(shape):
    print(f"The area is: {shape.area()}")

rect = Rectangle(4, 5)
circle = Circle(3)  
```

```python
print_area(rect)
print_area(circle)
```

Polymorphism with Abstract Base Classes

Abstract base classes are used to define common methods from a group of related object. they can enforce the derived classes implementent particular methods, promoting, consistency across different implementations

```python
# Polymorphism with Abstract Base Classes
from abc import ABC, abstractmethod
```

```python
# defined an abstract base class
class Vehicule(ABC):
    @abstractmethod
    def start_engine(self):
        pass

# derived class
class Car(Vehicule):
    def start_engine(self):
        #super().start_engine()
        print("The car engine has started.")  

class Motorcycle(Vehicule):
    def start_engine(self):
        #super().start_engine()
        print("The motorcycle engine has started.")

def start_vehicle_engine(vehicle):
    vehicle.start_engine()
    
```

```python
car = Car()
motorcycle = Motorcycle()
```

```python
start_vehicle_engine(car)
start_vehicle_engine(motorcycle)
```

# Encapsulation and Abstraction

Encapsulation is tha concept of wrapping data (variables) and methods (functions) together as a single unit. It restricts direct access to some of the object's components, wich is a means of prenventing accidental interference

```python
# Encapsulation with getters and setters methods
# public, private and protected attributes

class Person:    
    def __init__(self, name, age):
        self.name = name  # public attribute
        self._age = age   # protected attribute
        self.__ssn = "123-45-6789"  # private attribute

    def get_ssn(self):
        return self.__ssn
    
class Employee(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id
```

```python
person = Person("John Doe", 30)
print(person.name)  # Accessing public attribute
print(person._age)  # Accessing protected attribute (not recommended)   
print(person.get_ssn())  # Accessing private attribute through a getter method
```

```python
employee = Employee("Jane Smith", 28, "E12345")
print(employee.name)  # Accessing inherited public attribute
print(employee._age)  # Accessing inherited protected attribute
print(employee.employee_id)  # Accessing Employee-specific public attribute 
```

# Encapsulation with Getter and Setter

```python
class Person:    
    def __init__(self, name, age):
        self.__name = name # private attribute
        self.__age = age   # private attribute
        self.__ssn = "123-45-6789"  # private attribute

    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    def get_ssn(self):
        return self.__ssn
    
    def set_name(self, name):
        self.__name = name
    
    def set_age(self, age):
        self.__age = age
    
    def set_ssn(self, ssn):
        self.__ssn = ssn
```

```python
person2 = Person("Alice", 25)
print(person2.get_name())
print(person2.get_age())
```

```python
person2.set_age(26)
print(person2.get_age())
```

```python
person2.__age = 30
print(person2.get_age())
```

# Abstraction

Abstraction is the  concept of hiding the complex implementation details and showing only the necessary features of an object. It allows us to focus on what an object does rather than how it does it. 
In Python, we can achieve abstraction using abstract base classes (ABCs) from the `abc` module.

```python
from abc import ABC, abstractmethod
```

```python
class Vehicule(ABC):
    def drive(self):
        print("The vehicle is driving.")

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicule):
    def start_engine(self):
        print("The car engine has started.")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()
```

```python
car5 = Car()
operate_vehicle(car5)
```
