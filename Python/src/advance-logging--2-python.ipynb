# Logging with Multiple Loggers

```python
import logging
# create a logger for module 1
logger1 = logging.getLogger('module1')
logger1.setLevel(logging.DEBUG)

# create a logger for module 2
logger2 = logging.getLogger('module2')
logger2.setLevel(logging.INFO)

# Configure logging settings
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
# Log messages with different loggers
logger1.debug("This is a debug message from module 1")

logger2.warning("This is a warning message from module 2")
logger2.error("This is an error message from module 2")
```

# Loging in a real world example

```python
import logging
# Configure logging settings
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    handlers=[
        logging.FileHandler('./logs/app.log'), 
        logging.StreamHandler()
        ]
)

logger = logging.getLogger('my_app')
logger.info("Application started successfully")

def add(a, b):
    logger.debug(f"Adding {a} + {b}: {a+b}")
    return a + b

def subtract(a, b):
    logger.debug(f"Subtracting {b} - {a}: {a-b}")
    return a - b

def multiply(a, b):
    logger.debug(f"Multiplying {a} * {b}: {a*b}")
    return a * b

def divide(a, b):
    try:
        result = a / b
        logger.debug(f"Dividing {a} / {b}: {result}")
        return result
    except ZeroDivisionError:
        logger.error(f"Error dividing {a} by {b}: division by zero")
        return None

add(10, 5)
subtract(10, 5)
multiply(10, 5)
divide(10, 5)
divide(10, 0)
```

```python
# OPEN FILE AND READ ITS CONTENT
with open('./logs/app.log', 'r') as f:
    print(f.read())

    
```
