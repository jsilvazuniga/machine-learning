## Magic Methods

🧩 Object creation & representation
====================================

__init__(self, …) — Initializes the object after it is created.

__new__(cls, …) — Creates the object before __init__ runs.

__del__(self) — Called when the object is about to be destroyed.

__repr__(self) — Official string representation, used for debugging.

__str__(self) — Human‑friendly string representation (used by print).

__format__(self, format_spec) — Custom formatting with format() or f‑strings.


📦 Container behavior (list‑like or dict‑like)
===============================================
__len__(self) — Returns the number of elements.

__getitem__(self, key) — Access with obj[key].

__setitem__(self, key, value) — Assignment with obj[key] = value.

__delitem__(self, key) — Deletion with del obj[key].

__contains__(self, item) — Implements item in obj.

__iter__(self) — Returns an iterator.

__next__(self) — Returns the next element of an iterator.


➕ Arithmetic operators
=========================
__add__ — +

__sub__ — -

__mul__ — *

__truediv__ — /

__floordiv__ — //

__mod__ — %

__pow__ — **

```python
## Basic methods overwriting __str__
class Person:
    def __init__(self, name, age ):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name},{self.age} years old!"
    
    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"
    
person = Person("John", 30)
print(person)
print(repr(person))
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f"Point({self.x}, {self.y})"

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
p1 = Point(1, 2)
print(repr(p1))
```

## Custom Exception (Raise and Throw an exception)

```python
class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
year = int(input("Enter the dob: "))

age = 2024 - year

try:
    if age <= 30 and age >= 20:
        print("The age is valid so you can apply for the job!")
    else:
        raise dobException
except dobException:
    print("Sorry, you are not eligible for the job! ")
```

## Operator Overloading

```python
## Math Operation with vector
 
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)  
        
    def __mul__(self, other):
        return Vector(self.x * other.x, self.y * other.y)

    def __eq__(self, value):
        return self.x == value.x and self.y == value.y
    
    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

# create objects of the Vector class    
v1 = Vector(1, 2)   
v2 = Vector(3, 4)   

print(v1 + v2)  # Output: Vector(4, 6)
print(v1 - v2)  # Output: Vector(-2, -2)
print(v1 * v2)  # Output: Vector(3, 8)
print(v1 == v2) # Output: False

```

## Iterators

```python
my_list = [1, 2, 3]
print(my_list)
print(type(my_list))
```

```python
iterator=iter(my_list)
print(type(iterator))
print(iterator)
```

```python
# iterate over the list using the iterator
print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2
print(next(iterator))  # Output: 3
print(next(iterator))  # This will raise StopIteration since there are no more items to iterate over    
```

```python
iterator=iter(my_list)

try:
    print(next(iterator))  # Output: 1
    print(next(iterator))  # Output: 2
    print(next(iterator))  # Output: 3
    print(next(iterator))  # This will raise StopIteration since there are no more items to iterate over
except StopIteration:
    print("No more items to iterate over.")
```

## Generators

```python
def square(n):
    for i in range(n):
        yield i ** 2

for num in square(5):
    print(num)
    
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen = my_generator()
gen
```

```python
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2   
```

```python
for val in gen:
    print(val)
```

```python
# Practical example : Reading large files

def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

# Usage
for line in read_large_file('data/example.txt'):
    print(line)
```
