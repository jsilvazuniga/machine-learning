MLFlow Tracking Server

```python
import mlflow
```

```python
print("MLflow version: " + mlflow.__version__)
```

```python
# URL MLFLOW by default in this case is running in localhost
mlflow.set_tracking_uri("http://localhost:5000")
```

```python
mlflow.set_experiment("Check localhost connection")

with mlflow.start_run():
    mlflow.log_param("param1", 5)
    mlflow.log_metric("metric_test_1", 0.85)
    mlflow.log_metric("metric_test_2", 0.90)
    mlflow.log_metric("metric_test_3", 1.0)

                      
```
